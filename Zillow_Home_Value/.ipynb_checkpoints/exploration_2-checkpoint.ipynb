```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

color = sns.color_palette()
```

```python
properties = pd.read_csv('data/properties_2016.csv')
transaction = pd.read_csv('data/train_2016.csv', parse_dates=['transactiondate'])
```

```python
properties.rename(columns={'parcelid':'id_parcel', 'yearbuilt': 'build_year', 'basementsqft':'area_basement', 'yardbuildingsqft17':'area_patio',
                   'yardbuildingsqft26':'area_shed', 'poolsizesum':'area_pool', 'lotsizesquarefeet':'area_lot', 'garagetotalsqft':'area_garage',
                   'finishedfloor1squarefeet': 'area_firstfloor_finished', 'calculatedfinishedsquarefeet':'area_total_calc',
                   'finishedsquarefeet6':'area_base', 'finishedsquarefeet12':'area_live_finished','finishedsquarefeet13':'area_liveperi_finished',
                   'finishedsquarefeet15':'area_total_finished','finishedsquarefeet50':'area_unknown', 'unitcnt':'num_unit',
                   'numberofstories':'num_story', 'roomcnt':'num_room', 'bathroomcnt': 'num_bathroom', 'bedroomcnt':'num_bedroom',
                   'calculatedbathnbr':'num_bathroom_calc', 'fullbathcnt':'num_bath', 'threequarterbathnbr':'num_75_bath',
                   'fireplacecnt':'num_fireplace','poolcnt':'num_pool','garagecarcnt':'num_garage','regionidcounty':'region_county',
                   'regionidcity':'region_city', 'regionidzip':'region_zip', 'regionidneighborhood':'region_neighbor',
                   'taxvaluedollarcnt':'tax_total', 'structuretaxvaluedollarcnt':'tax_building', 'landtaxvaluedollarcnt':'tax_land',
                   'taxamount':'tax_property','assessmentyear':'tax_year','taxdelinquencyflag':'tax_delinquency',
                   'taxdelinquencyyear':'tax_delinquency_year','propertyzoningdesc':'zoning_property',
                   'propertylandusetypeid':'zoning_landuse', 'propertycountylandusecode':'zoning_landuse_county',
                   'fireplaceflag':'flag_fireplace','hashottuborspa':'flag_tub','buildingqualitytypeid':'quality',
                   'buildingclasstypeid':'framing', 'typeconstructiontypeid':'material','decktypeid':'deck',
                   'storytypeid':'story','heatingorsystemtypeid':'heating','airconditioningtypeid':'aircon',
                   'architecturalstyletypeid':'architectural_style'}, inplace=True)
```

```python
transaction.rename(columns={'parcelid':'id_parcel'}, inplace=True)
```

```python
properties.tax_delinquency = properties.tax_delinquency.map(lambda x: 1 if x=='Y' else 0)
properties.flag_tub = properties.flag_tub.map(lambda x: 1 if x==True else 0)
properties.flag_fireplace = properties.flag_fireplace.map(lambda x: 1 if x==True else 0)
```

### Transaction

```python
transaction.tail()
```

```python
plt.scatter(range(len(transaction)),np.sort(transaction['logerror'].values))
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.distplot(transaction['logerror'].values, bins=400, kde=False)
plt.xlim(-0.5,0.5)
plt.xlabel('logerror',fontsize=15)
plt.ylabel('Count', fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.distplot(np.abs(transaction['logerror'].values), bins=350, kde=False)
plt.xlim(-0.01, 0.5)
plt.xlabel('abs_logerror', fontsize=15)
plt.ylabel('count', fontsize=15)
plt.show()
```

```python
transaction['transaction_month'] = transaction['transactiondate'].dt.month
month_cnt = transaction['transaction_month'].value_counts()

plt.figure(figsize=(12,8))
sns.barplot(x = month_cnt.index, y=month_cnt.values, alpha=0.7, color=color[3])
plt.axvline(x= 9, color = 'black', lw=5, ls='--')
plt.xlabel('month')
plt.ylabel('count')
plt.show()
```

```python
mean_logerror = transaction[['transaction_month', 'logerror']].groupby(by='transaction_month').mean()
mean_abs_logerror = np.abs(transaction[['transaction_month', 'logerror']]).groupby(by='transaction_month').mean()

plt.subplot(2,1,1)
plt.plot(mean_logerror, 'ro-')
plt.xlabel('month')
plt.ylabel('mean_logerror')

plt.subplot(2,1,2)
plt.plot(mean_abs_logerror, 'bo-')
plt.xlabel('month')
plt.ylabel('mean_abs_logerror')
plt.show()
```

### Properties

```python
missing_df = properties.isnull().sum().reset_index()
missing_df.columns = ['column_name', 'missing_count']
missing_df = missing_df.ix[missing_df['missing_count']>0]
missing_df = missing_df.sort_values(by='missing_count')

ind = np.arange(missing_df.shape[0])
width= 0.9

fig, ax = plt.subplots(figsize=(12,18))
rects = ax.barh(ind, missing_df.missing_count.values, color='blue')
ax.set_yticks(ind)
ax.set_yticklabels(missing_df.column_name.values, rotation='horizontal')
ax.set_xlabel('Count of missing values')
ax.set_title('Number of missing values in each column')
plt.show()
```

```python
missing_df['missing_pct'] = missing_df['missing_count']/properties.shape[0]

fig, ax = plt.subplots(figsize=(12,18))
rects = ax.barh(ind, missing_df.missing_pct, color='green')
ax.set_yticks(ind)
ax.set_yticklabels(missing_df.column_name.values, rotation='horizontal')
ax.set_xlabel('Percentage of missing values')
ax.set_title('Percentage of missing values in each column')
plt.show()
```

```python
good_features = missing_df[missing_df.missing_pct<0.75]
good_features
```
